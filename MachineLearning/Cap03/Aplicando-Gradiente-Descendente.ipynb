```python
# Construindo o gráfico da função com x entre -15 e 5
# Construir vetor x com 1000 pontos entre -15 e 5
x = np.linspace(-15, 5, 1000)
y = np.power(x+5, 2)
plt.figure(figsize=(5, 4))
plt.plot(x, y, label='y = (x+5)^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função y = (x+5)^2')
plt.scatter(-5, 0, color='red', label='Mínimo da função', s=100)
plt.grid()
plt.legend()
plt.show()
```

## Aplicando o gradiente descendente

```python
# O algoritmo começa em x = 3
cur_x = 3

# Taxa de aprendizado (learning rate)
rate = 0.01

# Define quando parar o algoritmo
precision = 0.000001

# Inicializa o contador do passo anterior
previous_step_size = 1

# Número máximo de iterações
max_iters = 1000

# Contador de iterações
iters = 0

# Gradiente da função (derivada)
df = lambda x: 2*(x+5)
```

```python
while previous_step_size > precision or iters < max_iters:
    prev_x = cur_x # armazena o valor atual de x
    cur_x = cur_x - rate * df(prev_x) # gradiente descendente
    previous_step_size = abs(cur_x - prev_x) # variação do valor de x
    iters = iters+1 # incrementa o contador de iterações
    print('Iteração', iters, '- Valor de x é', cur_x)

print('O mínimo da função ocorre em:', cur_x)
```
