## Adição de Colunas e Índices COM e SEM Broadcasting

### Sem Broadcasting

```python
df['salario_final'] = 0 # Criando uma nova coluna com valores 0
df.head() # Visualizando o DataFrame
```

```python
df['salario_final'] = df['base_salary'] + df['guaranteed_compensation'] # Criando uma nova coluna com base em operações com outras colunas
df.head() # Visualizando o DataFrame
```

```python
df.insert(0, column='ID', value=range(1,1+len(df))) # Inserindo uma nova coluna no DataFrame. 
# loc é a posição, column é o nome da coluna e value é o valor da coluna
df.head() # Visualizando o DataFrame
```

```python
df.tail() # Visualizando o DataFrame
```

### Com Broadcasting

```python
df['base_salary'].add(5) # Adicionando 5 a cada valor da coluna base_salary
# não altera o DataFrame original
```

```python
df['base_salary'].head() # Visualizando a coluna base_salary
```

```python
df['base_salary'] = df['base_salary'].add(5) # Adicionando 5 a cada valor da coluna base_salary
df['base_salary'].head() # Visualizando a coluna base_salary
```

```python
df['base_salary_eur'] = df['base_salary'].mul(0.92) # Convertendo o salário base para euros
df.head() # Visualizando o DataFrame
```

```python
df.drop(columns=['base_salary_eur'], inplace=True) # Removendo a coluna base_salary_eur. inplace=True altera o DataFrame original
df.head() # Visualizando o DataFrame
```

```python
df.insert(6, column='base_salary_eur', value=df['base_salary'].mul(0.92)) # Inserindo a coluna base_salary_eur
df.head() # Visualizando o DataFrame
```

```python
df.rename(columns={'base_salary':'base_salary_usd',
                   'guaranteed_compensation': 'guaranteed_compensation_usd',
                   'salario_final': 'salario_final_usd'}, 
                   inplace=True) # Renomeando as colunas

df.head() # Visualizando o DataFrame
```

```python
df.set_index('ID', inplace=True) # Definindo a coluna ID como índice
df.head() # Visualizando o DataFrame
```
