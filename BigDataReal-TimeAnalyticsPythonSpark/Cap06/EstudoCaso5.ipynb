## Label Encoding

### Método 1

```python
df.prioridade_produto.value_counts()
```

```python
# criar dicionário para mapear a prioridade do produto:
dic_prioridade_produto = {'baixa': 1, 'media': 2, 'alta': 0}
dic_prioridade_produto
```

```python
df['prioridade_produto'] = df['prioridade_produto'].map(dic_prioridade_produto)
df.prioridade_produto.value_counts()
```

```python
df.modo_envio.value_counts()
```

```python
dic_modo_envio = {'Navio': 0, 'Aviao': 1, 'Caminhao': 2}
```

```python
df['modo_envio'] = df['modo_envio'].map(dic_modo_envio)
df.modo_envio.value_counts()
```

### Método 2

```python
df.genero.value_counts()
```

```python
labelEncoder_genero = LabelEncoder() # instanciando o objeto
labelEncoder_genero.fit(df['genero']) # ajustando o objeto
list(labelEncoder_genero.classes_) # classes mapeadas
```

```python
df.genero = labelEncoder_genero.transform(df['genero'])
df.genero.value_counts()
```

## One-Hot Encoding

```python
df.corredor_armazem.value_counts()
```

```python
df.performance_prioridade_envio.value_counts()
```

```python
df.performance_modo_envio.value_counts()
```

```python
df.faixa_desconto.value_counts()
```

```python
df.performance_faixa_desconto.value_counts()
```

```python
# Aplicar o One-Hot Encoding:
for cat in ['corredor_armazem', 
            'performance_prioridade_envio',
            'performance_modo_envio',
            'faixa_desconto',
            'performance_faixa_desconto']:
    oneHots = pd.get_dummies(df[cat], prefix=cat)
    # get_dummies cria um DataFrame com as variáveis dummies
    # prefix=cat adiciona o nome da categoria como prefixo
    df = df.join(oneHots)

df # visualizando o DataFrame
```

```python
df.drop(columns=['corredor_armazem',
                    'performance_prioridade_envio',
                    'performance_modo_envio',
                    'faixa_desconto',
                    'performance_faixa_desconto'], inplace=True)
df.drop(columns=['ID'], inplace=True)
df # visualizando o DataFrame
```

## Feature Scaling

### Normalização

```python
df.peso_gramas.sample(5)
```

```python
df.peso_gramas = MinMaxScaler().fit_transform(df[['peso_gramas']]) # normalizando a variável peso_gramas
```

```python
df.peso_gramas.sample(5)
```

```python
df.custo_produto.sample(5)
```

```python
df.custo_produto = MinMaxScaler().fit_transform(df[['custo_produto']]) # normalizando a variável custo_produto
```

```python
df.custo_produto.sample(5)
```

### Padronização

```python
df.desconto = StandardScaler().fit_transform(df[['desconto']]) # padronizando a variável desconto
df.desconto.sample(5)
```

```python
df.numero_chamadas_cliente = \
    StandardScaler().fit_transform(df[['numero_chamadas_cliente']]) # padronizando a variável numero_chamadas_cliente
df.numero_chamadas_cliente.sample(5)
```

```python
df.avaliacao_cliente = \
    StandardScaler().fit_transform(df[['avaliacao_cliente']]) # padronizando a variável avaliacao_cliente
df.avaliacao_cliente.sample(5)
```

```python
df.compras_anteriores = \
    StandardScaler().fit_transform(df[['compras_anteriores']]) # padronizando a variável compras_anteriores
df.compras_anteriores.sample(5)
```
