### Exercício 1 - Crie uma função Python usando somente operações matemáticas (não use outras funções ou métodos) que receba uma lista de números como argumento e devolva como resultado a média da lista de números.

```python
# Solução

# Lista de números
lista_numeros = [78, 80, 79, 80, 87, 85]

# Função
def calcula_media(lista):     
    
    # Coloque aqui sua solução
    media = 0
    for item in lista:
        media += item / len(lista)
    print(f'A média é: {media:.2f}')

# Chamada à função
calcula_media(lista_numeros)            
```

### Exercício 2 - Crie uma função Python que receba uma lista como argumento e retorne a mediana. A função deve retornar a mediana para lista com qualquer número de elementos. Use métodos auxiliares se necessário.

```python
# Solução

# Listas de números
lista_numeros_1 = [78, 80, 79, 81, 87, 85]
lista_numeros_2 = [78, 80, 79, 87, 85]

# Função
def calcula_mediana(nums):
    # Coloque aqui sua solução
    nums.sort()
    tamanho = len(nums)
    if tamanho % 2 == 0:
        mediana = (nums[tamanho // 2 - 1] + nums[tamanho // 2]) / 2
    else:
        mediana = nums[tamanho // 2]
    return mediana

# Chamada à função
resultado1 = calcula_mediana(lista_numeros_1)
resultado2 = calcula_mediana(lista_numeros_2)

print("A mediana da lista 1 é:", resultado1)
print("A mediana da lista 2 é:", resultado2)
```

### Exercício 3 - Considerando a lista de elementos abaixo que representa o número de produtos vendidos por dia em uma empresa, crie uma função Python que receba a lista e retorne a variância. Use métodos auxiliares se necessário.

```python
# Lista de números
unidades_vendidas = [5, 4, 3, 4, 8, 2, 5, 2, 5, 8, 10, 8, 7, 8, 7, 9, 10, 5, 7, 5, 7, 5, 4, 3, 5, 8, 4]
```

```python
# Solução

# Função 
def calcula_variancia(data):
    
    # Coloque aqui sua solução
    tamanho_amostra = len(data)
    media = sum(data) / tamanho_amostra
    total_dos_quadrados_das_diferencas = 0
    for dado in data:
        total_dos_quadrados_das_diferencas = total_dos_quadrados_das_diferencas + (dado - media) ** 2
    variancia = total_dos_quadrados_das_diferencas / (tamanho_amostra - 1)
    return variancia

# Chamada à função
v = round(calcula_variancia(unidades_vendidas), 2)

print("A variância da lista é:", v)
```

### Exercício 4 - Considerando a lista de elementos abaixo que representa o número de produtos vendidos por dia em uma empresa, crie uma função que calcule o desvio padrão. Desvio padrão é o quadrado da variância.

```python
# Lista de números
unidades_vendidas = [5, 4, 3, 4, 8, 2, 5, 2, 5, 8, 10, 8, 7, 8, 7, 9, 10, 5, 7, 5, 7, 5, 4, 3, 5, 8, 4]
```

```python
# Solução

# Import
import math

# Função
def calcula_std(dados):
    
    # Coloque aqui sua solução
    variancia = calcula_variancia(dados)
    desvio_padrao = math.sqrt(variancia)
    return desvio_padrao

# Chamada à função
s = round(calcula_std(unidades_vendidas), 2)
print("O desvio padrão da lista é:", s)
```

### Exercício 5 - Usando os pacotes NumPy e Scipy, calcule as estatísticas descritivas de tendência central e dispersão da lista abaixo.

```python
# Lista de números
unidades = [5, 4, 3, 5, 8, 2, 5, 2, 5, 8, 10, 8, 7, 9, 10, 5, 7, 5, 7, 5, 4, 3, 5, 8, 4]
```

```python
# Solução

import numpy as np
import scipy
from scipy import stats

# Coloque aqui sua solução

# Cálculo de assimetria (skewness)
assimetria = scipy.stats.skew(unidades)
print(f'A assimetria é: {assimetria:.2f}')

# Cálculo de curtose (kurtosis)
curtose = scipy.stats.kurtosis(unidades)
print(f'A curtose é: {curtose:.2f}')
```

**Análise de Assimetria:** como o valor de skewness calculado é maior que zero, a distribuição é assimétrica positiva. Isso significa que a cauda da distribuição está se estendendo para a direita.

**Análise de Curtose:** como o valor de kurtosis calculado é menor que zero, a distribuição é platicúrtica. Isso significa que a distribuição é mais achatada e tem caudas mais leves que a distribuição normal.

```python
# Gráfico de histograma
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(unidades, kde=True)
plt.title('Histograma das unidades vendidas')
plt.show()
```

```python
# Gráfico de densidade de probabilidade (tipo KDE) comparada a uma distribuição normal
sns.kdeplot(unidades, fill=True, alpha=0.5)
media_unidades = np.mean(unidades)
desvio_padrao_unidades = np.std(unidades)
gerar_dados = np.random.normal(media_unidades, desvio_padrao_unidades, 100000)
sns.kdeplot(gerar_dados, fill=True, alpha=0.1)
plt.xlabel('Unidades Vendidas')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição de Unidades Vendidas')
plt.xlim(0, 12)
plt.show()
```
